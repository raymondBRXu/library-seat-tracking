## 1. Introduction

Cornell FutureLibs is a web dashboard that shows real-time occupancy status of each library and each floor by analyzing library camera feeds with computer vision model to estimate the number of people in each floor. It also features a room booking function that allows student to plan group study in advance.

The motivation for this project comes from our experiences of tiring seat hunting in libraries, especially during peak times like final seasons, which often leads to wasted time and energy. FutureLibs dashboard provides one simple but helpful piece of information: which library and which floor is less crowded at this time.

## 2. Data

Our original intention was to analyze library camera feeds to estimate number of people in each floor. However, due to privacy issues, we are not able to obtain the real footage or stream source from Cornell Libraries. Also, parallel real-time processing of couple of dozens of streaming videos requires advanced hardware setup. Therefore, to demonstrate the idea, we used AI Image Generators (e.g. Google Nano Banana) to generate a series of images of camera footage frames for each floor, and iteratively analyze a batch of pictures which contains one frame from each floor to get the estimated number of people in each floor. The images are stored in `data/lib_images`.

Since each frame can only capture a part of each floor, we multiply the detected counts by 8 to get the displayed number of people on the dashboard to approximate the total number in that floor. When dashboard is launched, the `people_counter` starts analyzing images, saves the data in `library_occupancy.csv`. The `GUI/pages/1_Availability.py` page will then retrieves the most recent data from csv and display it in the dashboard.

## 3. Technical Stack

**Front End: `Streamlit`** Streamlit is a framework that allows use to build and share multi-page apps in pure python which is ideal for light-weight data dashboard design and iteration.

**Computer Vision and Object Detection: `OpenCV` and `Ultralytics YOLO`** This combination is widely-used for object detection tasks and performs well on tasks like people counting, makes it a good choice for occupancy estimation.

**Language:** `Python 3.13`

## 4. Future Works

In future developments, we plan to breaking each floor into smaller zones to display more detailed information. Also,fine-tuning a YOLO model on library scenes would improve detection accuracy.
